```python
def percentileID(sortTPM_index):# sampleTPM: PM1640protein, sortTPM_index: PM1640_sortindex
    percentiles = []
    for i in range(0,10):
        start = round(len(sortTPM_index) * i/10)
        end = round(len(sortTPM_index) * (i+1)/10) -1
        percentiles.append(sortTPM_index[start:end+1]) # percentile
    return percentiles
```

```python

from scipy.stats import spearmanr
from sklearn.metrics import mean_squared_error
import pandas as pd

epicseq1corr=[]
epicseq1mse=[]
for samplename in samples:
    outputmat = pd.read_csv(f'/athena/khuranalab/scratch/wel4007/notebooks/slurm/herberts/epicseq/1/{samplename}.withTPM.csv',sep='\t')[['TPM', 'inferredGEP']]
    outputmat = outputmat.dropna()
    print(outputmat.shape)

    # Spearman correlation
    corr, p_value = spearmanr(outputmat['TPM'], outputmat['inferredGEP'])
    epicseq1corr.append(corr)
    # Mean Squared Error
    mse = mean_squared_error(np.log2(outputmat['TPM']+1), outputmat['inferredGEP'])
    epicseq1mse.append(mse)
    print(f"{samplename}: Spearman={corr:.4f}, p={p_value:.4e}, MSE={mse:.4f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from scipy.stats import mannwhitneyu

# --- Rename columns ---
gt_res = gt_res.rename(columns={"log2TPM": "value"})
final_res = final_res.rename(columns={"correction": "value"})
epicseq_res = epicseq_res.rename(columns={"inferredGEP": "value"})
ulz_res = ulz_res.rename(columns={"sum_check": "value"})
pdx_res = pdx_res.rename(columns={"correction": "value"})

# --- Add dataset labels ---
gt_res["Dataset"] = "log2TPM"
final_res["Dataset"] = "cfOncoXpress"
epicseq_res["Dataset"] = "EPIC-seq"
ulz_res["Dataset"] = "Ulz"
pdx_res["Dataset"] = "train using PDX"

# --- Combine all datasets ---
combined_res = pd.concat([gt_res, final_res, epicseq_res, ulz_res, pdx_res], ignore_index=True)

# --- Plot setup ---
x = "Dataset"
y = "value"
hue = "class"

n_panels = 13
ncols = 4
nrows = int(np.ceil((n_panels + 1) / ncols))  # +1 for legend slot  still 3 rows

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(16, 9), constrained_layout=True)
axes = axes.flatten()

palette = {"CRPC-AR": "#8c6bb1", "CRPC-NE": "#d9d9d9"}

# --- Significance annotation helper ---
def get_sig_symbol(p):
    if p <= 0.0001:
        return "****"
    elif p <= 0.001:
        return "***"
    elif p <= 0.01:
        return "**"
    elif p <= 0.05:
        return "*"
    else:
        return None

datasets = ["log2TPM", "cfOncoXpress", "EPIC-seq", "Ulz", "train using PDX"]

# --- Plot each patient ---
for i, sample in enumerate(samples[:n_panels]):
    ax = axes[i]
    data_sample = combined_res[combined_res["Patient"] == sample]
    if data_sample.empty:
        fig.delaxes(ax)
        continue

    # Violin plot
    sns.violinplot(
        data=data_sample, x=x, y=y, hue=hue,
        dodge=True, palette=palette, inner=None, ax=ax
    )

    # Overlay dots
    sns.stripplot(
        data=data_sample, x=x, y=y, hue=hue,
        dodge=True, jitter=True, size=5, alpha=0.7,
        palette=palette, ax=ax
    )

    # Compute p-values and draw stars (no brackets)
    ymax_for_sample = data_sample[y].max()
    ymin_for_sample = data_sample[y].min()
    y_range = ymax_for_sample - ymin_for_sample
    top_y = ymax_for_sample

    for j, ds in enumerate(datasets):
        ar_vals = data_sample[(data_sample["Dataset"] == ds) & (data_sample[hue] == "CRPC-AR")][y]
        ne_vals = data_sample[(data_sample["Dataset"] == ds) & (data_sample[hue] == "CRPC-NE")][y]

        if len(ar_vals) > 0 and len(ne_vals) > 0:
            stat, p = mannwhitneyu(ar_vals, ne_vals, alternative='greater')
            sig_label = get_sig_symbol(p)

            if sig_label:
                y_max = max(ar_vals.max(), ne_vals.max())
                text_height = y_max + 0.05 * y_range
                ax.text(j, text_height, sig_label, ha='center', va='bottom', fontsize=10)
                if text_height > top_y:
                    top_y = text_height

    # Adjust y-limits so stars fit
    ax.set_ylim(ymin_for_sample, top_y + 0.15 * y_range)

    # Formatting
    ax.set_xlabel('')
    if i % ncols == 0:
        ax.set_ylabel('Expression')
    else:
        ax.set_ylabel('')

    ax.set_xticklabels(ax.get_xticklabels(), rotation=30, ha='right')

    sample_new = sample.replace('-', '').replace('BaselinecfDNA', 'B').replace('ProgressioncfDNA', 'P')
    ax.set_title(sample_new, fontsize=12)
    ax.tick_params(axis='y', labelsize=12)
    ax.get_legend().remove()

# --- Legend on the last empty subplot ---
legend_ax = axes[n_panels]  # use the next empty cell
legend_ax.axis("off")  # no axes

# Create a clean legend in that subplot
handles, labels = axes[0].get_legend_handles_labels()
legend_ax.legend(handles[:2], labels[:2],
                 loc='center', fontsize=12, title='Class',
                 ncol=1, frameon=False)

# --- Remove remaining unused subplots (if any) ---
for j in range(n_panels + 1, len(axes)):
    fig.delaxes(axes[j])

# --- Save and show ---
# plt.savefig('2025figures/Fig5_sigdiff.pdf', dpi=500, bbox_inches='tight')
plt.show()
```

```python
import pathlib
import re
import pandas as pd
import matplotlib.pyplot as plt

# 1. Set directory
directory = pathlib.Path("/athena/khuranalab/scratch/wel4007/notebooks/slurm/herberts/downsample/epicseqcontrol01_bam/test/")  

data = []

# 2. Read all *.out files
for file in directory.glob("*.out"):
    with open(file) as f:
        lines = f.read().strip().splitlines()
        if len(lines) < 3:
            continue
        spearman_line = lines[-3]
        mse_line = lines[-2]

        # 3. Extract cov and tf from filename
        match = match = re.search(r"cov(\d+(?:\.\d+)?)_tf(\d+(?:\.\d+)?)", file.name)

        
        if not match:
            continue
        cov = float(match.group(1))
        tf = float(match.group(2))
        # print(cov, tf)
        # 4. Extract values
        # Regex: optional sign, digits, optional decimal; then ; then std
        spearman_match = re.search(r"spearman:\s*([-+]?\d*\.?\d+)\s*\s*([-+]?\d*\.?\d+)", spearman_line, re.I)
        mse_match = re.search(r"mse:\s*([-+]?\d*\.?\d+)\s*\s*([-+]?\d*\.?\d+)", mse_line, re.I)
        
        if spearman_match and mse_match:
            spearman = float(spearman_match.group(1))
            spearman_std = float(spearman_match.group(2))
            mse = float(mse_match.group(1))
            mse_std = float(mse_match.group(2))
        
            data.append((cov, tf, spearman, spearman_std, mse, mse_std))

# 5. Build dataframe
df = pd.DataFrame(data, columns=["cov", "tf", "spearman","spearman_std", "mse","mse_std"])
df = df.sort_values(["cov", "tf"])


```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_metric(metric, ylabel, threshold):
    plt.figure(figsize=(10, 6))
    
    std_col = f"{metric}_std"
    unique_covs = sorted(df["cov"].unique())
    unique_tfs = sorted(df["tf"].unique())
    
    n_covs = len(unique_covs)
    n_tfs = len(unique_tfs)
    
    bar_width = 0.8 / n_covs
    x = np.arange(n_tfs)

    # Use a colormap with enough distinct colors (e.g., 'tab20')
    cmap = plt.cm.get_cmap('tab20', n_covs)
    
    for i, cov in enumerate(unique_covs):
        group = df[df["cov"] == cov].set_index("tf").reindex(unique_tfs)
        x_pos = x + i * bar_width
        
        plt.bar(
            x_pos,
            group[metric].values,
            width=bar_width,
            yerr=group[std_col].values,
            capsize=3,
            color=cmap(i),   # Use unique color from colormap
            label=f"{cov:.3f}x"
        )

    plt.xticks(x + bar_width*(n_covs - 1)/2, [f"{t:.2f}" for t in unique_tfs])
    plt.axhline(y=threshold, color='black', linestyle='--', linewidth=1)

    plt.xlabel("Tumor Fraction", fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    plt.legend(title="Sequencing depth", fontsize=12, bbox_to_anchor=(1.02, 1), loc='upper left')
    plt.tight_layout()
    # plt.savefig('2025figures/downsample_mse.pdf', dpi=500, bbox_inches='tight')

    plt.show()
```

```python
plot_metric("mse", "Mean Squared Error", 4.404)#
```
